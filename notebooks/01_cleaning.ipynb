```python
import pandas as pd
import numpy as np
import sqlite3
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the dataset
spotify_data = pd.read_csv('../data/raw/spotify_data_12_20_2023.csv', low_memory=False)
```

```python
# looking at how many rows and columns in the dataset, and how many unique rows of data there are
print(spotify_data.shape)
print(spotify_data['track_id'].nunique())
```

```python
remove_cols = ['analysis_url', 'track_href', 'uri', 'type', 'rn','duration_ms','genre_0', 'genre_1', 'genre_2', 'genre_3', 'genre_4']

spotify_data = spotify_data.drop(columns=remove_cols)
```

```python
print("\nData Types:\n")
print(spotify_data.dtypes)
```

```python
# fixing the data types so they're optimised for BI

# Convert release_date to datetime
spotify_data['release_date'] = pd.to_datetime(spotify_data['release_date'], errors='coerce')

# Clean artist_genres
spotify_data['artist_genres'] = (spotify_data['artist_genres'].str.replace(r"[\[\]']", "", regex=True).str.strip())

# Replace empty genre strings with NaN
spotify_data['artist_genres'] = spotify_data['artist_genres'].replace('', pd.NA)

print("\nData Types:\n")
print(spotify_data.dtypes)
```

```python
# final clean dataset preview
spotify_data.head()
```

```python
spotify_data.tail()
```

```python
# checking how many null values we have in the columns
nulls = spotify_data.isnull().sum()
print(nulls[nulls > 0])
```

```python
# filling null values
spotify_data['artist_genres'] = spotify_data['artist_genres'].fillna('Unknown')
```

```python
# dropping null rows
audio_features = ['acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'mode', 'speechiness', 'tempo', 'time_signature', 'valence', 'key']

spotify_data = spotify_data.dropna(subset=audio_features)
print("rows after dropping missing audio features:", spotify_data.shape)
```

```python
# checking how many null values we have in the columns
nulls = spotify_data.isnull().sum()
print(nulls[nulls > 0])
```

```python
# EDA/data checking
print("Rows and Columns:", spotify_data.shape)
print("\nExplicit breakdown:\n", spotify_data['explicit'].value_counts())
print("\nAlbum types:\n", spotify_data['album_type'].value_counts())
print("\nRelease years:\n", spotify_data['release_year'].value_counts().sort_index().tail(10))
print("\nTop 10 genres:\n", spotify_data['artist_genres'].value_counts().head(10))
```

```python
# because the genre data is written weirdly, we must save 2 versions:
# the original version (above), and the new version which focuses on the genres.
```

```python
spotify_clean = spotify_data.copy()
spotify_clean.to_csv('../data/cleaned/spotify_cleaned.csv', index=False)
print("Rows and Columns:", spotify_clean.shape)
```

```python
# this version uses .exploded from pandas to seperate a block of data from the col/row

spotify_genres = spotify_data.copy()

# removes square brackets and quoation marks from a string
spotify_genres['artist_genres'] = spotify_genres['artist_genres'].str.replace(r"[\[\]']", "", regex=True).str.split(',')

# explodes/seperates the column
spotify_genres = spotify_genres.explode('artist_genres')

spotify_genres['artist_genres'] = spotify_genres['artist_genres'].str.strip().replace('', 'Unknown').fillna('Unknown')

spotify_genres.to_csv('../data/cleaned/spotify_genres.csv', index=False)
print("Rows and Columns:", spotify_genres.shape)
```
